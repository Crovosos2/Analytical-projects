```python
query_2 = '''
SELECT b.title, b.book_id, ra.avg_rating, rev.total_review
FROM books b
LEFT JOIN (SELECT book_id, COUNT(review_id) AS total_review
    FROM reviews
    GROUP BY book_id) AS rev ON b.book_id=rev.book_id
LEFT JOIN (SELECT book_id, AVG(rating) AS avg_rating
    FROM ratings
    GROUP BY book_id) AS ra ON b.book_id=ra.book_id
'''

request2 = pd.io.sql.read_sql(query_2, con = engine)

display(request2)
#WITH
#rev AS (SELECT book_id, COUNT(review_id) AS total_review
#    FROM reviews
#    GROUP BY book_id),
#ra AS (SELECT book_id, AVG(rating) AS avg_rating
#    FROM ratings
#    GROUP BY book_id)
    
#SELECT b.title, b.book_id, ra.avg_rating, rev.total_review
#FROM books b
#LEFT JOIN ra ON b.book_id=ra.book_id
#LEFT JOIN rev ON b.book_id=rev.book_id
```

## Задание 3.
Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query_3 = '''
WITH
book AS (SELECT books.publisher_id AS publisher_id, COUNT(books.book_id) AS total_book, publishers.publisher AS publisher
    FROM books
    LEFT JOIN publishers ON books.publisher_id=publishers.publisher_id
    WHERE num_pages > 50
    GROUP BY 1, 3
    ORDER BY total_book DESC
    LIMIT 1)
    
SELECT book.publisher_id, book.publisher, book.total_book
FROM book
'''

request3 = pd.io.sql.read_sql(query_3, con = engine)

display(request3)
print('Издательство {} выпустило больше всего книг толще 50 страниц, а именно {}шт.'.format(request3.loc[0][1],
                                                                                         request3.loc[0][2]))
```

## Задание 4.
Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query_4 = '''
SELECT t.author, AVG(t.avg_rating) AS avg_rating
FROM (SELECT a.author AS author, AVG(r.rating) AS avg_rating
    FROM books b
    LEFT JOIN authors a ON b.author_id = a.author_id
    LEFT JOIN ratings r ON b.book_id = r.book_id
    GROUP BY b.book_id, a.author_id
    HAVING COUNT(r.rating_id) >=50) AS t 
GROUP BY t.author
ORDER BY avg_rating DESC
LIMIT 1
'''

request4 = pd.io.sql.read_sql(query_4, con = engine)

display(request4)
print('Автор {} имеет среднюю оценку своих книг равную {}'.format(request4.loc[0][0], request4.loc[0][1]))
```

## Задача 5.
Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query_5 = '''
WITH
t AS (SELECT username, COUNT(review_id) AS count_rew
    FROM reviews
    WHERE username IN (SELECT username
        FROM ratings
        GROUP BY username
        HAVING COUNT(rating_id) > 50)
        GROUP BY username)
    
SELECT AVG(count_rew) AS avg_rew
FROM t
'''

request5 = pd.io.sql.read_sql(query_5, con = engine)
display(request5)
display('Среднее количество обзоров, которые поставили пользователи с количеством оценок от 50 - {}'.format(request5.loc[0][0]))
```
